```python
import re
#a pipeline to clean up text 
def tweet_cleaner(text):
    #remove RT
    text = re.sub("RT @[\w]*:","",text)
    #remove twitter handles (@user)
    text = re.sub("(@[A-Za-z0-9_]+)","", text) 
    #remove url links
    text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\(\),]|'\
                  '(?:%[0-9a-fA-F][0-9a-fA-F]))+','', text) 
    # remove whitespaces 
    text = ' '.join(text.split()) 
    return text
```

```python
#Apply text cleaning function to cleaned_text column in df_tweets dataframe
df_tweets['text'] = df_tweets['text'].apply(lambda x: tweet_cleaner(x))
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sid = SentimentIntensityAnalyzer()
```

```python
df_tweets['scores'] = df_tweets['text'].apply(lambda text: sid.polarity_scores(text))

df_tweets.head(10)
```

```python
def sentimentPredict(sentiment):
  if sentiment['compound'] >= 0.05:
    return "Positive"
  elif sentiment['compound'] <= -0.05:
    return "Negative"
  else:
    return "Neutral"
```

```python
df_tweets['label'] =df_tweets['scores'].apply(lambda x: sentimentPredict(x))
```
